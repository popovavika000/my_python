```python
import pandas as pd

# Загрузка данных
lectures_df = pd.read_csv('lectures.csv')
questions_df = pd.read_csv('questions.csv')

# Просмотр первых нескольких строк
print(lectures_df.head())
print(questions_df.head())




```

```python
# Информация о лекциях
print(lectures_df.info())

# Информация о вопросах
print(questions_df.info())
```

```python
# Количество уникальных лекций
print("Уникальные лекции:", lectures_df['lecture_id'].nunique())

# Количество уникальных вопросов
print("Уникальные вопросы:", questions_df['question_id'].nunique())

# Уникальные категории для лекций и вопросов
print("Уникальные части лекций:", lectures_df['part'].unique())
print("Уникальные части вопросов:", questions_df['part'].unique())
```

```python
# Уникальные теги лекций
print("Уникальные теги лекций:", lectures_df['tag'].nunique())

# Уникальные теги вопросов
tags_questions = questions_df['tags'].str.split().explode().unique()
print("Уникальные теги вопросов:", len(tags_questions))

# Распределение тегов для вопросов
tag_counts = questions_df['tags'].str.split().explode().value_counts()
print("Теги для вопросов (топ 10):")
print(tag_counts.head(10))
```

```python
# Распределение типов лекций
lecture_type_counts = lectures_df['type_of'].value_counts()
print("Типы лекций и их количество:")
print(lecture_type_counts)
```

```python
# Объединение данных по разделам (part) для лекций и вопросов
lecture_part_merged = lectures_df[['lecture_id', 'part']].drop_duplicates()
question_part_merged = questions_df[['question_id', 'part']].drop_duplicates()

# Слияние данных по разделу (part)
merged_df = pd.merge(lecture_part_merged, question_part_merged, on='part', how='inner')
print("Число совпадений между лекциями и вопросами по разделам:", merged_df.shape[0])
print(merged_df.head())
```

```python
# Обзор лекций
print("Обзор лекций")
print(lectures_df.describe())
print("\n")

# Обзор вопросов
print("Обзор вопросов")
print(questions_df.describe())
```

```python
# Уникальные значения для столбца `part` и `type_of` (лекции)
print("Уникальные разделы для лекций:")
print(lectures_df['part'].value_counts())

print("\nТипы лекций:")
print(lectures_df['type_of'].value_counts())
```

```python
# Уникальные теги для вопросов
tags_questions = questions_df['tags'].str.split().explode().unique()
print("Уникальные теги вопросов:")
print(tags_questions)

# Подсчет, сколько раз встречается каждый тег
tag_counts = questions_df['tags'].str.split().explode().value_counts()
print("\nТеги для вопросов (топ 10):")
print(tag_counts.head(10))
```

```python
# Соединение лекций и вопросов по разделам (`part`)
lecture_part_merged = lectures_df[['lecture_id', 'part']].drop_duplicates()
question_part_merged = questions_df[['question_id', 'part']].drop_duplicates()

# Слияние
merged_df = pd.merge(lecture_part_merged, question_part_merged, on='part', how='inner')
print("Число совпадений между лекциями и вопросами по разделам:", merged_df.shape[0])
print(merged_df.head())
```

```python
# Распределение разделов для лекций и вопросов
print("Распределение разделов для лекций:")
print(lectures_df['part'].value_counts())

print("\nРаспределение разделов для вопросов:")
print(questions_df['part'].value_counts())
```

```python
# Анализ переходов между лекциями и вопросами по разделам
# Можно рассматривать как последовательность, если есть временные метки или явная связь между лекциями и вопросами.
lectures_df['next_lecture_id'] = lectures_df['lecture_id'].shift(-1)
questions_df['next_question_id'] = questions_df['question_id'].shift(-1)

print(lectures_df[['lecture_id', 'part', 'next_lecture_id']].head())
print(questions_df[['question_id', 'part', 'next_question_id']].head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans

# Преобразуем теги в признаки для кластеризации
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(questions_df['tags'].fillna(''))

# Кластеризация вопросов
kmeans = KMeans(n_clusters=5, random_state=42)
questions_df['cluster'] = kmeans.fit_predict(X)

# Выводим распределение кластеров
cluster_counts = questions_df['cluster'].value_counts()
print(cluster_counts)
```
